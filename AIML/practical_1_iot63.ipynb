```python
#sheikhwasimuddin iot63
print("TWO ROOM VACUUM CLEANER AGENT")
rooms = {
    "A": "dirty",
    "B": "dirty"
}

current_room = "A"
def vacuum_agent(room, status):

    if status == "dirty":
        return "clean"
    else:
        return "move"

while True:

    print("\nAgent is in Room:", current_room)
    print("Room Status:", rooms[current_room])

    action = vacuum_agent(current_room, rooms[current_room])

    if action == "clean":
        print("Action: Cleaning the room")
        rooms[current_room] = "clean"
    elif action == "move":
        print("Action: Moving to another room")

        if current_room == "A":
            current_room = "B"
        else:
            current_room = "A"
    if rooms["A"] == "clean" and rooms["B"] == "clean":
        print("\nBoth rooms are clean!")
        break
```

```python
#sheikh wasimuddin iot-63
from collections import deque
initial_state = [
    [1, 2, 3],
    [4, 0, 6],
    [7, 5, 8]
]

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def generate_states(state):
    new_states = []
    x, y = find_blank(state)
    moves = [
        (-1, 0),
        (1, 0),
        (0, -1),
        (0, 1)
    ]
    for dx, dy in moves:
        nx = x + dx
        ny = y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            new_states.append(new_state)
    return new_states

def convert_to_tuple(state):
    return tuple(tuple(row) for row in state)
def bfs(start):
    visited = set()
    queue = deque([(start, [])])
    while queue:
        state, path = queue.popleft()
        if state == goal_state:
            return path + [state]
        visited.add(convert_to_tuple(state))
        for next_state in generate_states(state):
            if convert_to_tuple(next_state) not in visited:
                queue.append((next_state, path + [state]))
    return None
def print_state(state):
    for row in state:
        print(row)
    print()
solution = bfs(initial_state)
if solution:
    print("Solution Found!\n")
    step = 0
    for state in solution:
        print("Step", step)
        print_state(state)
        step += 1
else:
    print("No solution found.")
```
